```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as s
df = pd.read_csv("C:/Users/Manas Tiwari/Desktop/Zomato data .csv")
```

# checking is there any null value present in data

```python
df.isnull().sum()
```

# change rate column in folat and split by ( / )


```python
df['rating_value'] = df['rate'].str.split('/').str[0].astype(float)
```

# Q.1 - what type of resturant do the majority of customer order from  ??


```python
a = s.countplot(df,x = "listed_in(type)")
for bars in a.containers:
    a.bar_label(bars)
```

conclusion : as we can see from the above graph that maximum customer order from dining

# Q.2 : how many votes has each type of restaurant recived from customer  ??

```python
gp = df.groupby('listed_in(type)')['votes'].sum()
plt.pie(gp,labels = gp.index,autopct = "%1.2f%%")
plt.show()
```

conclusion : as we can see from the above graph that maximum votes recived to dining..

# Q.3 : what are the rating that majority restaurant have recived ??

```python
a = s.countplot(df, x= "rating_value")
for bars in a.containers:
    a.bar_label(bars)
```

#conclusion : as we can see from the above graph that majority restaurants recived rating of 3.8

# Q.4 : what is the average price that a couple spent in resturants ??

```python
a = s.countplot(df,x = "approx_cost(for two people)")
for bars in a.containers:
    a.bar_label(bars)
```

as we can see the average spending amount is 300 by a couple
